```julia
using Plots
using DataFrames
gr()
```

```julia
function babble_sort(arglist::Array{T, 1}, log=false::Bool)::Array{T, 1} where T<:Number
  list = arglist |> copy
  len = arglist |> length
  
  for i=1:len
    for j=1:(len-i)
      left = list[j]
      right = list[j+1]
      if log
        "Order: $i:$j...($left):($right) compare:$(left<right)" |> println
      end
      if left>right
        list[j], list[j+1] = list[j+1], list[j]
      end
    end
  end
  
  return list
end
```

```julia
# Defied in previous Cell.
# ?babble_sort
```

```julia
N = 1000

list = rand(N) .|> x->round(x, digits=2)

"Data initialized. length: $(length(list)), max: $(maximum(list)), min: $(minimum(list))" |> println

babblesort = babble_sort
@time sorted_list = babblesort(list)

sort(list) == sorted_list ? "Sort success" : "Sort failure"
```
